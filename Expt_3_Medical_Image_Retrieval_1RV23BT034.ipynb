```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define the base directory of the unzipped dataset
base_dir = '/content/AugmentedAlzheimerDataset'

# Get the list of classification folders
class_folders = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]

print("Found classification folders:", class_folders)

# Display 2 images from each class
for class_folder in class_folders:
    class_path = os.path.join(base_dir, class_folder)
    image_files = [f for f in os.listdir(class_path) if f.endswith('.jpg') or f.endswith('.png')]

    print(f"\nDisplaying images for class: {class_folder}")
    plt.figure(figsize=(10, 5))
    for i in range(min(2, len(image_files))): # Display up to 2 images
        img_path = os.path.join(class_path, image_files[i])
        img = mpimg.imread(img_path)
        plt.subplot(1, 2, i + 1)
        plt.imshow(img)
        plt.title(image_files[i])
        plt.axis('off')
    plt.show()
```

From A Website

```python
import requests
import matplotlib.pyplot as plt
from PIL import Image
from io import BytesIO

image_url = 'https://prod-images-static.radiopaedia.org/images/3060519/downloaded_image20240531-964-3xflgv_big_gallery.jpeg'
image_title = 'MRI Image from Radiopaedia'

try:
    response = requests.get(image_url)
    response.raise_for_status() # Raise an exception for bad status codes (4xx or 5xx)

    img = Image.open(BytesIO(response.content))

    plt.imshow(img)
    plt.title(image_title)
    plt.axis('off')
    plt.show()

except requests.exceptions.RequestException as e:
    print(f"Error downloading image: {e}")
except Exception as e:
    print(f"Error processing or displaying image: {e}")
```
